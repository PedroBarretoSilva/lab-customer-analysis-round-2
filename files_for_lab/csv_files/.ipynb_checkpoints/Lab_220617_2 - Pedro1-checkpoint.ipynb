```python
# the graph above has too many unique values
# we have to Group total_claim_amount into bins
# to check which would be the best bins, weÂ´ll calculate max, min, mean
```

```python
print("Maximum value :", basedados4["total_claim_amount"].max())
print("Minimum value :", basedados4["total_claim_amount"].min())
print("Mean value :", basedados4["total_claim_amount"].mean())
```

```python
# Function to define the bins

def tca_bins(x):
    if x<=450:
        return 1
    elif x<=900:
        return 2
    elif x<=1350:
        return 3
    elif x<=1800:
        return 4
    else :
        return 5
    

basedados4["total_claim_amount_bins"]=list(map(tca_bins,basedados4["total_claim_amount"]))
basedados4
    
```

```python
basedados4.response_rate.value_counts()
sns.barplot(x="total_claim_amount_bins", y="response_rate", data=basedados4)
plt.show()
```

```python
# 6 - Show a plot of the response rate by  income
# the problem of many unique values will take place again, so weÂ´ve repeated the same approach :
```

```python
print("Maximum value :", basedados4["income"].max())
print("Minimum value :", basedados4["income"].min())
print("Mean value :", basedados4["income"].mean())
```

```python
# Function to define the bins

def income_bins(x):
    if x<=20000:
        return 1
    elif x<=40000:
        return 2
    elif x<=60000:
        return 3
    elif x<=80000:
        return 4
    else :
        return 5
    

basedados4["income_bins"]=list(map(income_bins,basedados4["income"]))
basedados4
```

```python
basedados4.response_rate.value_counts()
sns.barplot(x="income_bins", y="response_rate", data=basedados4)
plt.show()
```

# Customer Analysis Round 6

```python
#Enconding the Categorical variables so that we convert it into numbers and become able to use them in a regression model

#checkinf the status of the variable X_cat:
```

```python


X_cat1=X_cat.drop("customer", axis=1)   #dropping the columns customer ID, as its not likely to be relevant for our analysis

display(X_cat1.head())
print(X_cat1.shape)
```

```python
#WeÂ´ll enconde these categorical variables by using the One Hot methodology
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder=OneHotEncoder(drop="first").fit(X_cat1)
cols=encoder.get_feature_names(input_features=X_cat1.columns)
X_cat1_encoded=pd.DataFrame(encoder.transform(X_cat1).toarray(),columns=cols)
X_cat1_encoded.head(20)
```

```python
# Now that the categorical variables have a numberic shape, weÂ´ll build the final X matrix of explanatory variables (X_final) 
# to be used in our regression. 
# the X_final matrix weÂ´ll put together  categorical and numeric variables
# WeÂ´ve chosen to work with the numeric variables Transformed via the MinMax methodology
```

```python
X_final=pd.concat([X_cat1_encoded,X_num_scale],axis=1)
```

```python
# making the Linear regression Model based on the matrix X_final
```

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score



lm=linear_model.LinearRegression()
lm.fit(X_final,y)
print(lm.intercept_)
print(lm.coef_)

```

```python
# assessing the quality of the model

r2_score(lm.predict(X_final),y)
```

# Train-test split

```python
# Making the Train-test split of data, taking 20% for test
# weÂ´ll build the model again using only 80% of the historical data (X_train) and keeping 20% aside (X_test) to test the model 
# as if it was new data for which we donÂ´t know the y (variable we are trying to forecast)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X_final, y, test_size=0.20, random_state=42)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

# Linear Regression Model (Train)

```python
# Doing again the Linear Regression Model (II), but now taking only 80% of the sample
```

```python
lm2=linear_model.LinearRegression()
lm2.fit(X_train,y_train)
print(lm.intercept_)
print(lm.coef_)
```

```python
# assessing the quality of the model II (for 80% of the  sample):

r2_score(lm2.predict(X_train),y_train)
```

# Test and Metrics

```python
# testing the model now by simulating a  prediction for the variable y, based on the 20% of explanatory variables
# we are basically pretending that we donÂ´t know the y (target variable) for those 20% . By doing so, 
# weÂ´ll be able to assess the ModelÂ´s Quality (by calculating metrics like R2, MSE, MAE etc...) under a scenario where
# the model is actualy being aplied to new data (data which have not been used to build the model) in order to make predictions
# since it's "more difficult to predict the future than the past", the R2 and other Quality metrics become much more meanfull  
# vs the R2 calculted while using 80% of data (during the Train stage)
```

```python
y_predict=lm2.predict(X_test)
```

```python
# errors observed based on the test :
```

```python
teste_erros=y_predict-y_test
```

```python
# Model's Outcome Summary :
```

```python
outcome=pd.DataFrame({"predicted":y_predict, "actual":y_test, "difference": teste_erros })
outcome
```

```python
# these differeces reflect the Quality / Accurrancy that are shown on the four main metrics below:
```

# Metrics 

```python
# R2

r2_score(y_predict,y_test) # the R2 is showing a better value vs the achieved in the lm and lm2
```

```python
# MSE = MEAN SQUARED ERROR

mse=mean_squared_error(y_predict,y_test) 
mse
```

```python
# RMSE = ROOT MEAN SQUARED ERROR

import math
rmse=math.sqrt(mse)
rmse
```

```python
from sklearn.metrics import mean_absolute_error

# MAE = MEAN ABSOLUTE ERROR

mae = mean_absolute_error(y_predict,y_test) 
mae
```
