```python
#step4.1: Analysing the column "state" e "response"

basedados3["state"].value_counts(dropna=False)
```

```python
# both "state" and "response" have the same frequency and they are not much (5,78%), but are they related?

# letÂ´s make a filter by "state" with Nan, to see if it hapens on the same rows where "response" e having also Nan


basedados3[basedados3["state"].isna()]

```

```python

# as we can see, both columns have Nan on the same rows and they are not much (less than 6%), so its a good idea to drop
#its lines with "Nan", as weÂ´ll clean 2 columns without loosiung too  much information
```

```python
basedados4=basedados3.dropna(subset=["state"]).copy()
basedados4
```

```python
basedados4.shape


```

```python
#step4.2: Analysing the columns "months since last claim" e  "number_of_open_complaints"

basedados4["months_since_last_claim"].value_counts(dropna=False)
```

```python
basedados4["months_since_last_claim"].shape
```

```python
basedados4["number_of_open_complaints"].value_counts(dropna=False)
```

```python
basedados4["number_of_open_complaints"].shape
```

```python
# these two columns have the same number of Nan, but IÂ´ve decided to sort them out base on two different assumptions:

# months_since_last_claim : I assumed that there were zero claims
# number_of_open_complaints : I replaced the Nan by the mean of the column
```

```python
# months_since_last_claim : making the replacement and confirming it has been done successfully


basedados4["months_since_last_claim"]=basedados4["months_since_last_claim"].fillna(0)

print(basedados4["months_since_last_claim"].value_counts(dropna=False))

basedados4["months_since_last_claim"].value_counts(dropna=False).sum()


```

```python
# number_of_open_complaints : making the replacement and confirming it has been done successfully


number_of_open_complaints_media=basedados4["number_of_open_complaints"].mean()
print(number_of_open_complaints_media)

basedados4["number_of_open_complaints"]=basedados4["number_of_open_complaints"].fillna(number_of_open_complaints_media)


print(basedados4["number_of_open_complaints"].value_counts(dropna=False))

basedados4["number_of_open_complaints"].value_counts(dropna=False).sum()




```

```python
#step4.3: Analysing the column "vehicle_type"
```

```python
basedados4["vehicle_type"].value_counts(dropna=False)
```

```python
basedados4.shape
```

```python
# vehicle_type : I replaced the Nan by "M" (Manual) 
#vehicle_type : making the replacement and confirming it has been done successfully
```

```python
basedados4["vehicle_type"]=basedados4["vehicle_type"].fillna("M")

print(basedados4["vehicle_type"].value_counts(dropna=False))

basedados4["vehicle_type"].value_counts(dropna=False).sum()
```

```python
# checking the remaining Nan values


basedados4.isna().sum()
```
